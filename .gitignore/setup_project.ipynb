Generating synthetic sales data

```python
np.random.seed(42)
n = 1000

df = pd.DataFrame({
    "month":np.tile(np.arange(1,13),n//12+1)[:n],
    "advertising_spend":np.random.uniform(1000, 10000,n),
    "discount_pct":np.random.uniform(0,30,n),
    "num_salespeople":np.random.randint(5,50,n),
    "competitor_price":np.random.uniform(50,200,n),
    "is_holiday":np.random.randint(0,2,n)
})

df["sales"] = (
    500
    +3.5*df["advertising_spend"]
    -20 * df["discount_pct"]
    +150 * df["num_salespeople"]
    -2.5 * df["competitor_price"]
    +800 * df["is_holiday"]
    +np.random.normal(0,2000,n)
).clip(0)

df.to_csv("data/sales_data.csv",index=False)
print(df.describe())
print("\nshape:",df.shape)
print("saved -> d")
```
