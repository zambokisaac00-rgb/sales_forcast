```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/sales_data.csv")
print("shape:",df.shape)
print("\nFirst 5 rows:\n",df.head())
print("\nMissing values:\n",df.isnull().sum())
print("\nStates:\n",df.describe())
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["sales"],bins=40,kde=True,color="steelblue")
plt.title("sales_distribution")
plt.xlabel("sales")
plt.tight_layout()
plt.show()
```

```python
features = ["advertising_spend","discount_pct","num_salespeople","competitor_price"]
fig,axes = plt.subplots(2,2, figsize=(12,8))
for ax ,feat in zip(axes.flatten(),features):
    ax.scatter(df[feat],df["sales"],alpha=0.3,color="steelblue")
    ax.set_xlabel(feat)
    ax.set_ylabel("sales")
    ax.set_title(f"sales vs {feat}")
plt.tight_layout()
plt.show()
```
